```python
#Import all libraries
# for EDA
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
# plots appear inside the notebook

# scikit learn models
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.ensemble import RandomForestClassifier

# for model evaluation
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("Credit Card Defaulter Prediction.csv")
df
```

```python
df.info()
```

```python
df.MARRIAGE.unique()
```

```python
df.isna().sum()
```

```python
df= df.rename(columns=str.lower)
```

```python
df.head()
```

```python
df.drop(['id'],axis=1,inplace=True)
df
```

```python
sex = pd.get_dummies(df.sex)
sex.drop(['F'],axis=1,inplace=True)
sex
```

```python
marriage = pd.get_dummies(df.marriage)
marriage.drop(['0'],axis=1,inplace=True)
marriage
```

```python
df.marriage.value_counts()
```

```python
education = pd.get_dummies(df.education)
education.drop(['0'],axis=1,inplace=True)
education
```

```python
df.drop(['sex','marriage','education'],axis=1,inplace=True)
df = pd.concat([df,sex,marriage,education],axis=1)
df
```

```python
#Seperate label and feature
x = df.drop("default ",axis=1)
y = df['default ']
```

```python
x.head()
```

```python
y.head()
```

```python
target = pd.get_dummies(y)
target
```

```python
y = target.drop(['N'],axis=1)
```

```python
y
```

```python
y = y['Y']
y.head()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,
                                                y,
                                                test_size = 0.2)
```

```python
from sklearn import svm
clf = svm.SVC()
clf.fit(x_train,y_train)
```

```python
clf.score(x_test,y_test)
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(x_train,y_train)
```

```python
gnb.score(x_test,y_test)
```
